### 8. For Loops

```python
total = 0

for value in [3, 2, 19]:
    total += value
    
print('Total is: {0}'.format(total))
```

```python
total = 0

for value in [2, 3, 19]:
    for weight in [3, 2, 1]:
        total += value * weight

print('Total is: {0}'.format(total))
```

```python
total = 0

for value, weight in zip([2,3,19], [0.2,0.3,0.5]):
    total += weight * value
    
print('Weighted average is: {0}'.format(total))
```

```python
rng = np.random.default_rng(1)

A = rng.standard_normal((127, 5))

A
```

```python
M = rng.choice([0, np.nan], p=[0.8,0.2], size=A.shape)

M
```

```python
A += M

A
```

```python
D = pd.DataFrame(
    A, 
    columns=[
        'food',
        'bar',
        'pickle',
        'snack',
        'popcorn'
    ]
)

D[:3]
```

```python
for col in D.columns:
    template = 'Column "{0}" has {1:.2%} missing values'

    print(
        template.format(
            col,
            np.isnan(D[col]).mean()
        )
    )
```
