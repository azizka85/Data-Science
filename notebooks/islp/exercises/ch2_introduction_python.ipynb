### 8. Investigate College data

```python
import pandas as pd

college = pd.read_csv('../data/College.csv')

college
```

```python
college2 = pd.read_csv('../data/College.csv', index_col=0)

college2
```

```python
college3 = college.rename({'Unnamed: 0': 'College'}, axis=1)

college3 = college3.set_index('College')

college3
```

```python
college = college3

college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']]);
```

```python
college.boxplot('Outstate', by='Private')
```

```python
college['Elite'] = pd.cut(
    college['Top10perc'],
    [0, 0.5, 1],
    labels=['No', 'Yes']
)

college['Elite'].value_counts()
```

```python
college.boxplot('Outstate', by='Elite')
```

### 9. Investigating Auto data

```python
auto = pd.read_csv(
    '../data/Auto.data', 
    delim_whitespace=True,
    na_values=['?']
)

auto
```

```python
auto.isna().sum()
```

```python
auto2 = auto.dropna()

auto2.isna().sum()
```

```python
auto2
```

```python
auto2.describe()
```

```python
auto3 = auto2.drop(auto2.index[9:85], axis=0)

auto3.describe()
```

```python
pd.plotting.scatter_matrix(auto2);
```

```python
import matplotlib.pyplot as plt

target_column = 'mpg'

columns = auto2.columns.drop(target_column)
fig, axes = plt.subplots(1, len(columns), figsize=(15, 4))

for i, column in enumerate(columns):
    ax = axes[i]
    auto2.plot.scatter(x=column, y=target_column, ax=ax)
    ax.set_title(f'{target_column} vs {column}')
    ax.grid(True)

plt.show()
```

### 10. Investigating Boston data

```python
from ISLP import load_data

boston = load_data('Boston')

boston
```

```python
boston.describe()
```

```python
pd.plotting.scatter_matrix(boston[['crim', 'nox', 'age', 'dis']]);
```

```python
crim_mean = boston['crim'].mean()
high_crim_index = boston['crim'] > crim_mean

boston[high_crim_index].describe()
```

```python
tax_mean = boston['tax'].mean()
high_tax_index = boston['tax'] > tax_mean

boston[high_tax_index].describe()
```

```python
ptratio_mean = boston['ptratio'].mean()
high_ptratio_index = boston['ptratio'] > ptratio_mean

boston[high_ptratio_index].describe()
```
